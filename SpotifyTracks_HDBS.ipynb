```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from hdbscan import HDBSCAN
```

```python
file_path = "dataset.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
df.info()
```

```python
df.head()
```

```python
df = df[['danceability','energy','loudness','speechiness','acousticness','instrumentalness','liveness','valence','tempo']]
df = df.dropna()
```

```python
X = StandardScaler().fit_transform(df)
```

```python
hdbs = HDBSCAN(min_samples=1,min_cluster_size=30,metric='euclidean',cluster_selection_epsilon=0.5)
```

```python
df['cluster'] = hdbs.fit_predict(X)
```

```python
df['cluster'].value_counts()
```

```python
sns.scatterplot(data=df,x='energy',y='acousticness',hue='cluster',s=10)
plt.show()
```
